## Color transparency

```python
red = Image.open('red_color.jpg')
```

```python
purple = Image.open('purple.png')
```

```python
red
```

```python
red.putalpha(128)
```

```python
red
```

```python
purple.paste(im=red,box=(0,0),mask=red)
```

```python
purple
```

```python
purple.save('pink.png')
```
